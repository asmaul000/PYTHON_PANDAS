Clean and Transform Data

```python
# Fill missing genres with 'Unknown'
df['genres'] = df['genres'].fillna('Unknown')
```

```python
# Fill missing values. Only fill object type columns with "unknown".
for col in df.columns:
    if df[col].dtype == 'object':
        df[col].fillna("unknown", inplace=True)

# Verify that there are no more null values in object columns
print("Null values after filling object columns:")
print(df.isnull().sum())
```

```python
# Save the modified DataFrame to a CSV file
df.to_csv('modified_netflix_data.csv', index=False)
# Download the saved CSV file
files.download('modified_netflix_data.csv')
```

```python
# Create a new column for the first listed genre
df['main_genre'] = df['genres'].apply(lambda x: x.split(',')[0] if pd.notnull(x) else 'Unknown')
df.head()
```

```python
# Save the modified DataFrame to a CSV file
df.to_csv('modified_netflix_data.csv', index=False)
# Download the saved CSV file
files.download('modified_netflix_data.csv')
```

```python
# Check if a title is available in Bangladesh
df['is_available_in_bd'] = df['availableCountries'].str.contains("Bangladesh", na=False)
display(df.head())
```

```python
# Content released over the years
release_trend = df.groupby('releaseYear').size()
release_trend
```

```python
# Optional: plot the trend (if using matplotlib or seaborn)
import matplotlib.pyplot as plt

release_trend.plot(kind='line', figsize=(10, 5), title='Content Released Over Years')
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.show()
```

```python
# plot trend using year and main genres
import matplotlib.pyplot as plt
# Analyze trend using year and main genres
genre_year_trend = df.groupby(['releaseYear', 'main_genre']).size().unstack(fill_value=0)

# Plot the trend for the top N main genres (adjust N as needed)
top_genres = df['main_genre'].value_counts().head(10).index
genre_year_trend[top_genres].plot(kind='line', figsize=(15, 8), title='Trend of Top Main Genres Over Years')
plt.xlabel("Year")
plt.ylabel("Number of Titles")
plt.grid(True)
plt.legend(title='Main Genre')
plt.show()
```

```python
#heatmap year and genres main

import matplotlib.pyplot as plt
import seaborn as sns

# Analyze trend using year and main genres for heatmap
genre_year_heatmap_data = df.groupby(['releaseYear', 'main_genre']).size().unstack(fill_value=0)

# Select the top N main genres for the heatmap (adjust N as needed)
top_genres_heatmap = df['main_genre'].value_counts().head(15).index
genre_year_heatmap_subset = genre_year_heatmap_data[top_genres_heatmap]

# Create the heatmap
plt.figure(figsize=(15, 10))
sns.heatmap(genre_year_heatmap_subset.T, cmap="YlGnBu", linewidths=.5)
plt.title('Heatmap of Content Release by Year and Main Genre')
plt.xlabel("Release Year")
plt.ylabel("Main Genre")
plt.show()
```

```python
# prompt: Analyze which genres are most popular in each year

import matplotlib.pyplot as plt
import seaborn as sns
# Analyze which genres are most popular in each year
# To do this, we can group by releaseYear and main_genre and count the occurrences
genre_popularity_by_year = df.groupby(['releaseYear', 'main_genre']).size().reset_index(name='count')

# Find the most popular genre for each year
most_popular_genres_yearly = genre_popularity_by_year.loc[genre_popularity_by_year.groupby('releaseYear')['count'].idxmax()]

print("\nMost Popular Genres by Year:")
print(most_popular_genres_yearly)

# Filter out rows where 'releaseYear' is 'unknown' before plotting
most_popular_genres_yearly_filtered = most_popular_genres_yearly[most_popular_genres_yearly['releaseYear'] != 'unknown']


# You can also visualize this trend
plt.figure(figsize=(15, 8))
sns.lineplot(data=most_popular_genres_yearly_filtered, x='releaseYear', y='count', hue='main_genre')
plt.title('Trend of Most Popular Genres Over Years')
plt.xlabel("Release Year")
plt.ylabel("Number of Titles (Most Popular Genre)")
plt.grid(True)
plt.legend(title='Most Popular Genre', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# prompt: Find content available in Bangladesh with a high IMDb rating

# Filter content available in Bangladesh and with high IMDb rating
# Define a threshold for "high IMDb rating" (you can adjust this)
imdb_rating_threshold = 7.5

bangladesh_high_rating_content = df[
    (df['is_available_in_bd']) &
    (df['imdbAverageRating'] >= imdb_rating_threshold)
].sort_values(by='imdbAverageRating', ascending=False)

# Display the results
print(f"\nContent available in Bangladesh with an IMDb rating of {imdb_rating_threshold} or higher:")
display(bangladesh_high_rating_content[['title', 'type', 'imdbAverageRating', 'genres']].head(20)) # Displaying top 20
```
